# Diabetes Diagnosis Based on some dependencies with Artificial Neural Network

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

# Importing Dataset

```python
dataset = pd.read_csv('datasets_228_482_diabetes.csv')
dataset
```

# Data Preprocessing

## Splitting into features and target variable

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Taking care of Missing Values

```python
y= y.reshape(len(y),1)
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values= np.nan, strategy = 'mean')
imputer_y = SimpleImputer(missing_values= np.nan, strategy = 'mean')
imputer.fit(X[:,0:])
imputer_y.fit(y)
X[:,0:] = imputer.transform(X[:,0:])
y = imputer_y.transform(y)
```

## Splitting into Training and Test Set

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,stratify=y,random_state = 0)
```

## Scaling the training set

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
X_train[:, 0:] = sc_x.fit_transform(X_train[:, 0:])
```

# Artificial Neural Network

## Adding Input Layer, Hidden Layer, Output Layer, Compiling the layers and fitting the training set in the model

```python
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units = 8, input_shape=(8,), activation='relu'))
ann.add(tf.keras.layers.Dense(units = 8, activation='relu'))
ann.add(tf.keras.layers.Dense(units = 8, activation='relu'))
ann.add(tf.keras.layers.Dense(units = 8, activation='relu'))
ann.add(tf.keras.layers.Dense(units = 8, activation='relu'))
ann.add(tf.keras.layers.Dense(units = 8, activation='relu'))
ann.add(tf.keras.layers.Dense(units = 8, activation='relu'))
ann.add(tf.keras.layers.Dense(units = 8, activation='relu'))
ann.add(tf.keras.layers.Dense(units = 1, activation='sigmoid'))
ann.compile(optimizer='adam' , loss='binary_crossentropy' , metrics=['accuracy'])
ann.fit(X_train, y_train, batch_size=128, epochs=1500)
```

# Testing the Test set

## Scaling the Test set

```python
X_test[:, 0:] = sc_x.transform(X_test[:,0:])
```

## Predicting the result from scaled test set and visualising both Predicted values and actual values side by side

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred>0.5)
np.set_printoptions(precision = 2)
print(np.concatenate((y_pred.reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), 1))
```

## Visualising accuracy of Prediction with the help of Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

## Accuracy of the model

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

## Predicting a single set of values

```python
print(ann.predict(sc_x.transform([[6,148,72,35,0,33.6,0.627,50]])) > 0.5)
```
